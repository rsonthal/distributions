```julia
using Plots, Interact, Knet
include("./../GAN/dcgan2.jl")
```

```julia
function glayer1(x0, w, m; training=true)
    x = w[1] * x0 
    x = batchnorm(x, m, w[2]; training=training)
    x = relu.(x)
    return x
end

function glayer2(x0, w, m; training=true)
    x = deconv4(w[1], x0; stride=2)
    x = batchnorm(x, m, w[2]; training=training)
    return x
end

function glayer3(x0, w, m; training=true)
    x = deconv4(w[1], x0)
    x = batchnorm(x, m, w[2]; training=training)
    x = relu.(x)
    return x
end
```

```julia
wd, wg, md, m = DCGAN.main("--epochs 0 --seed 1 --loadfile /Users/rishi/Project/GAN/full/models/0054.jld2");
```

```julia
z = DCGAN.sample_noise(Array{Float64},100,9);
```

```julia
training = false
x1 = glayer1(z, wg[1:2], m[1]; training=training)
x2 = glayer1(x1, wg[3:4], m[2]; training=training)
x3 = reshape(x2, 4,4,50,size(x1,2))
@manipulate for f = 1:50
    p = [heatmap(x3[:,:,f,i]', yflip = true, color=:grays) for i=1:9]
    plot(p..., layout = 9)
end
```

```julia
x4 = glayer2(x3, wg[5:6], m[3]; training=training)
@manipulate for f = 1:50
    p = [heatmap(x4[:,:,f,i]', yflip = true, color=:grays) for i=1:9]
    plot(p..., layout = 9)
end
```

```julia
x5 = glayer3(x4, wg[7:8], m[4]; training=training)
@manipulate for f = 1:20
    p = [heatmap(x5[:,:,f,i]', yflip = true, color=:grays) for i=1:9]
    plot(p..., layout = 9)
end
```

```julia
x6 = glayer2(x5, wg[9:10], m[5]; training=training)
@manipulate for f = 1:20
    p = [heatmap(x6[:,:,f,i]', yflip = true, color=:grays) for i=1:9]
    plot(p..., layout = 9)
end
```

```julia
x7 = tanh.(deconv4(wg[end-1], x6) .+ wg[end])
p = [heatmap(x7[:,:,1,i]', yflip = true, color=:grays) for i=1:9]
plot(p..., layout = 9)
```
